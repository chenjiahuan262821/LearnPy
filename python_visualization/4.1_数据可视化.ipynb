#### 2.2 数据选择及可视化

```python
data = np.random.standard_normal((10000, 2))
```

```python
df = pd.DataFrame(data, columns=['Data1', 'Data2'])
df.head()
```

```python
import matplotlib as mpl
mpl.rcParams['font.sans-serif'] = ['SimHei']    # 指定默认字体
mpl.rcParams['axes.unicode_minus'] = False     # 解决保存图像是负号'-'显示为方块的问题
df[(df.Data1 > 2) & (df.Data2 < -2)].plot(x='Data1', y='Data2', kind='scatter',title = 'Outlier图')       #scatter是plot里面的内置参数；
```

### 3. Matplotlib图形绘制

#### 3.1 Matplotlib基础图形绘制

**Plt常用功能：**

--plt.plot() : 绘制图形；x轴和y轴；
    
--plt.show()：图形显示；IDE,pycharm是一定要加这句话的；
    
--plt.figure()：生成图形对象，确定图片大小；
    
--plt.title()：确定图片标题；
    
--plt.xlabel()：确定图片x轴的名字；
    
--plt.legend()：显示图例；
    
--plt.scatter()：绘制散点图；
    
--plt.grid(True)：出现网格；
    
--plt.subplot()：绘制子图；
    
--plt.bar()：绘制柱状图；

#### 3.1.1 添加图例和设置线型及颜色

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot([1,2,3,4],[2,4,6,8],lw = 1.0)
plt.ylabel('Numbers')
plt.xlabel('Numbers')
# plt.show()
```

```python
import seaborn as sns
plt.plot([1,2,3,4],[2,4,6,8], 'r--',linewidth=1.0, label = 'Line',)    #'ro',plt.ylim(0,10)
plt.ylabel('Numbers')
plt.legend()     #显示图例；
plt.ylim(0,10)
# plt.show()
```

#### 3.1.4 分别编写绘图命令，输出在同一张图内

```python
from matplotlib import style
style.use('ggplot')           #matplotlib下的一种style；

x = [1,5,10]
y = [2,8,16]

x2 = [10,5,1]
y2 = [9,20,7]

# 分别编写绘图命令，输出在同一张图内
plt.figure(figsize = (8,6))    #生成图片对象；
plt.plot(x,y,linewidth=1, label = 'line1')     #这里的label是图例；
plt.plot(x2,y2,linewidth=1, label = 'line2')

plt.title('Two Line Plot')
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.legend()
```

**3.3 Matplotlib两维图形绘制**

```python
np.random.seed(2000)
y = np.random.randn(20, 2).cumsum(axis=0)
y
```

```python
plt.figure(figsize=(7, 4))
plt.plot(y, lw=1.5)

plt.grid(True)
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('2 dimensions Plot')
```

```python
plt.figure(figsize=(7, 4))
plt.plot(y[:, 0], lw=1.5, label='line1')
plt.plot(y[:, 1], lw=1.5, label='line2')

plt.grid(True)
plt.legend(loc=0)
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('2 dimensions Plot')
```

```python
y[:, 0] = y[:, 0] * 100

plt.figure(figsize=(8, 6))
plt.plot(y[:, 0], lw=1.5, label='Line1')
plt.plot(y[:, 1], lw=1.5, label='Line2')

plt.grid(True)
plt.legend()
plt.xlabel('index')
plt.ylabel('value')
plt.title('2 dimensions Plot')
```

```python
fig1,ax = plt.subplots(figsize = (10,6))      #plt.subplots会返回一个figure对象和一个坐标轴对象；

plt.plot(y[:, 0], lw=1.5, label='line1')
plt.legend(loc=8)
plt.xlabel('x')
plt.ylabel('y1')
plt.title('plot')

ax2 = ax.twinx()       #复制上一张子图的横坐标轴；
plt.plot(y[:, 1], 'g', lw=1.5, label='line2')
plt.legend(loc=10)
plt.ylabel('y2')
```

#### 3.4 Matplotlib子图绘制

```python
plt.figure(figsize=(8,6))
plt.subplot(211)
plt.plot(y[:, 0], lw=1.5, label='1st')
plt.legend(loc=0)
plt.ylabel('Value')
plt.title('Sub Plot')

plt.subplot(212)
plt.plot(y[:, 1], 'm', lw=1.5, label='2nd')
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
```
